### Calculate variance and mean of the dataset

```python
def calculate_normalization(dataloader):
    mean = torch.zeros(3)
    m2 = torch.zeros(3)
    n = len(dataloader)
    for images, labels in dataloader:
        mean += images.mean([0, 2, 3]) / n
        m2 += (images ** 2).mean([0, 2, 3]) / n
    var = m2 - mean ** 2
    return mean, var.sqrt()
```

```python
mean, std = calculate_normalization(train_loader)
print('Mean: ',mean,' - STD: ',std)
```

### Linear Projection of Flattened Patches

In this section, we split an image into fixed-size patches, linearly embed each of them, add position embeddings, and feed the resulting sequence of vectors to a standard Transformer encoder. We at first explain the details of this component by running each code on a sample input, and at the end, we will define the class.

```python
from IPython.display import Image
Image(filename='pics/linear_projection.png',width=700) 
```

##### Sample batch of input images
$x \in \mathbb{R}^{B\times C\times W\times H}$

```python
x = torch.randn(B, C, W, H)
```

##### Generate blocks from Input
Each flattened 2D patches is $x_p \in \mathbb{R}^{B\times N\times (P^2\times C)}$. `Unfold` function here extracts flatten blocks from a batched input tensor. Our kernel size is $P$, and stride is $P$.

```python
unfold = nn.Unfold(kernel_size=(P, P),stride=P)
x_blocks = unfold(x).transpose(1,2)
x_blocks.shape
```

##### Define matrix $E \in \mathbb{R}^{(P^2\times C)\times D}$

```python
E = torch.randn(P*P*C, D)
E.size()
```

##### Define matrix $E_{\text{pos}} \in \mathbb{R}^{(N+1)\times D}$

```python
E_pos = torch.randn(N+1, D)
E_pos.size()
```

##### Define a learnable embedding for classes
We should define an embedding for classes, but because class embedding must not be different for each image, we expand this embedding for each image in the batch. So the final embedding for classes is $x_{\text{class}} \in \mathbb{R}^{B\times 1\times D}$.

```python
x_class = torch.zeros(1, 1, D)
x_class = x_class.expand(B,-1,-1)
x_class.size()
```

##### The input of the Transformer encoder 

$z_0 = [x_{\text{class}};x_p E]+E_{\text{pos}}, \quad z_0 \in \mathbb{R}^{B\times (N+1)\times D}$.

```python
z_0 = torch.cat((x_class , torch.matmul(x_blocks, E)),dim=1) + E_pos
z_0.size()
```

##### Now, we put things together and define `Embedding` class of our model.

We also add a dropout to calculate output of this module

```python
class Embedding(nn.Module):
    
    def __init__(self, D, P, N, C, B,dropout_rate):
        super(Embedding, self).__init__()
        
        self.unfold = nn.Unfold(kernel_size=(P, P),stride=P)
        self.E = nn.Parameter(torch.randn(P*P*C, D))
        self.E_pos = nn.Parameter(torch.randn(N, D))
        self.x_class = nn.Parameter(torch.zeros(1, 1, D))
        self.dropout = nn.Dropout(dropout_rate)
        
        
    def forward(self,x):
        x_class = self.x_class.expand(x.shape[0],-1,-1)
        x_blocks = self.unfold(x).transpose(1,2)
        x = torch.cat((x_class , torch.matmul(x_blocks, self.E)),dim=1) + self.E_pos
        x = self.dropout(x)
        
        return x
```

### Transformer Encoder

We can split Transformer Encoder into two sections. In following sections, we will explore layers of multiheaded selfattention (MSA) and then a multilayer perceptron (MLP) with one hidden layer.

```python
from IPython.display import Image
Image(filename='pics/transformer_encoder.png', width = 200) 
```

#### MSA

At first, we should do layer norm (LN) on the input. Then, in MSA section, .  At the end, we have residual connections after every MSA.

$z_{l}^{\prime} = \text{MSA}(\text{LN}(z_{l-1}))+z_{l-1}, \quad l=1,\dots,L$

##### Layer Norm
We calculate $z_0$ from the past section, we define a layer norm function and then apply it over the input.

$\text{LN}(z_0) \in \mathbb{R}^{B\times (N+1)\times D}$.

```python
layer_norm = nn.LayerNorm(D)
ln_z_0 = layer_norm(z_0) 
ln_z_0.size()
```

##### Define $U_{qkv}$

In the paper [[1]](#Refrences), they define it as $U_{qkv} \in \mathbb{R}^{D \times 3D_h}$, but because we want to do multiplication $k$ number of $U_{qkv}$ on input, we define it as $U_{qkv} \in \mathbb{R}^{D \times (K \times 3D_h)}$.

```python
U_qkv = torch.randn(D, K * 3 * D_h)
U_qkv.size()
```

##### Calculate $\text{LN}(z_0)U_{qkv}$
$\text{LN}(z_0)_{i}$ for each batch is:
$$
\text{LN}(z_0)_{i} = 
\begin{pmatrix}
l_{0,0} & l_{0,1} & \dots & l_{0,D-1}\\ 
l_{1,0} & l_{1,1} & \dots & l_{1,D-1} \\
\vdots & \vdots & \ddots &\vdots \\
l_{N,0} & l_{N,1} & \dots & l_{N,D-1}
\end{pmatrix},\quad i=0,\dots,B-1
$$

Also, we defined $U_{qkv}$ as below later:
$$
U_{qkv} = \begin{pmatrix}
u_{0,0} & u_{0,1} & \dots & u_{0,K \times 3D_h-1}\\ 
u_{1,0} & u_{1,1} & \dots & u_{1,K \times 3D_h-1} \\
\vdots & \vdots & \ddots &\vdots \\
u_{D-1,0} & u_{D-1,1} & \dots & u_{D-1,K \times 3D_h-1}
\end{pmatrix}
$$

Therefore,
$$
\text{LN}(z_0)_{i}U_{qkv} = \begin{pmatrix}
a_{0,0} & a_{0,1} & \dots & a_{0,K \times 3D_h-1}\\ 
a_{1,0} & a_{1,1} & \dots & a_{1,K \times 3D_h-1} \\
\vdots & \vdots & \ddots &\vdots \\
a_{N,0} & a_{N,1} & \dots & a_{N,K \times 3D_h-1}
\end{pmatrix}, \quad a_{k,j} = l_{k,0}u_{0,j} + l_{k,1}u_{1,j} + \dots +  l_{k,D-1}u_{N,j}
$$

where $k=0,1,\dots,N$, and $j=0,1,\dots,K \times 3D_h-1$. The final matrix will be:

$$\text{LN}(z_0)U_{qkv} \in \mathbb{R}^{B \times (N+1) \times (K \times 3D_h)}$$


```python
tmp = torch.matmul(ln_z_0, U_qkv)
tmp.size()
```

##### Reshape the past matrix

Because we changed the dimensions of $U_{qkv}$, we need to reshape $\text{LN}(z_0)_{i}U_{qkv}$ for each element of the batch from $(N+1) \times (K \times 3D_h)$ to $(N+1) \times K \times 3D_h$ as below:

$$
(\text{LN}(z_0)_{i}U_{qkv})_{\text{reshape}} = \begin{pmatrix}
A_{0}\\ 
A_{1} \\
\vdots \\
A_{N}
\end{pmatrix}, \quad A_{k} =  \begin{pmatrix}
a_{k,0} & a_{k,1} & \dots & a_{k,3D_h-1}\\ 
a_{k,3D_h} & a_{k,3D_h+1} & \dots & a_{k,2\times3D_h-1} \\
\vdots & \vdots & \ddots &\vdots \\
a_{k,(K-1)3D_h} & a_{k,(K-1)3D_h+1} & \dots & a_{k,K \times 3D_h-1}
\end{pmatrix}, \quad k = 0,\dots,N
$$

And then, we need to transpose dimention of 1 and 2 (from $(N+1) \times K \times 3D_h$ to $K \times (N+1) \times 3D_h$). Then we will have this matrix:

$$
((\text{LN}(z_0)_{i}U_{qkv})_{\text{reshape}})_{\text{transpose}} = \begin{pmatrix}
B_{0}\\ 
B_{1} \\
\vdots \\
B_{K-1}
\end{pmatrix}, \quad B_{j} =  \begin{pmatrix}
a_{0,j\times 3D_h} & a_{0,j\times 3D_h+1} & \dots & a_{0,(j+1)\times 3D_h}\\ 
a_{1,j\times 3D_h} & a_{1,j\times 3D_h+1} & \dots & a_{1,(j+1)\times 3D_h} \\
\vdots & \vdots & \ddots &\vdots \\
a_{N,j\times 3D_h} & a_{N,j\times 3D_h+1} & \dots & a_{N,(j+1)\times 3D_h}
\end{pmatrix}, \quad j = 0,\dots,K-1
$$

We will do above equations for each batch, then our final matrix will be:

$$
((\text{LN}(z_0)_{i}U_{qkv})_{\text{reshape}})_{\text{transpose}}  \in \mathbb{R}^{B \times K \times (N+1) \times 3D_h}
$$

```python
tmp= tmp.reshape(B,N+1,K,3*D_h).transpose(1,2)
tmp.size()
```

##### Split matrix to $q$,$k$, and $v$

In this section, we split each $B_j$, and the size of a single chunk is $D_h$. Each chunk is a view of the original tensor. Thus, from each $B_j$, we have:

$$
q_j = \begin{pmatrix}
a_{0,j\times D_h} & a_{0,j\times D_h+1} & \dots & a_{0,(j+1)\times D_h}\\ 
a_{1,j\times D_h} & a_{1,j\times D_h+1} & \dots & a_{1,(j+1)\times D_h} \\
\vdots & \vdots & \ddots &\vdots \\
a_{N,j\times D_h} & a_{N,j\times D_h+1} & \dots & a_{N,(j+1)\times D_h}
\end{pmatrix}
$$

$$
k_j = \begin{pmatrix}
a_{0,(j+1)\times D_h} & a_{0,(j+1)\times D_h+1} & \dots & a_{0,(j+2)\times D_h}\\ 
a_{1,(j+1)\times D_h} & a_{1,(j+1)\times D_h+1} & \dots & a_{1,(j+2)\times D_h} \\
\vdots & \vdots & \ddots &\vdots \\
a_{N,(j+1)\times D_h} & a_{N,(j+1)\times D_h+1} & \dots & a_{N,(j+2)\times D_h}
\end{pmatrix}
$$

$$
v_j = \begin{pmatrix}
a_{0,(j+2)\times D_h} & a_{0,(j+2)\times D_h+1} & \dots & a_{0,(j+3)\times D_h}\\ 
a_{1,(j+2)\times D_h} & a_{1,(j+2)\times D_h+1} & \dots & a_{1,(j+3)\times D_h} \\
\vdots & \vdots & \ddots &\vdots \\
a_{N,(j+2)\times D_h} & a_{N,(j+2)\times D_h+1} & \dots & a_{N,(j+3)\times D_h}
\end{pmatrix}
$$

Thus, $q_j$, $k_j$, and $v_j \in \mathbb{R}^{(N+1) \times D_h}$, and $q$, $k$, and $v \in \mathbb{R}^{B \times K \times (N+1) \times D_h}$.

```python
q,k,v = torch.split(tmp, D_h, dim=3)
q.size()
```

##### Calculate $A$ and apply softmax over it.

Matrix $A$ calculated as below:

$$
A = \text{softmax}(\dfrac{qk^{T}}{\sqrt{D_h}})
$$

$$
A \in \mathbb{R}^{B \times K \times (N+1) \times (N+1)}
$$

```python
softmax = torch.nn.Softmax(dim=-1)
A = softmax(torch.div(torch.matmul(q, k.transpose(2,3)),D_h ** -0.5))
A.size()
```

##### Calculate matrix k number of $Av$

$Av \in \mathbb{R}^{B \times K \times (N+1) \times D_h}$

```python
SA = torch.matmul(A, v)
SA.size()
```

##### Define matrix $U_{msa}$

$U_{msa} \in \mathbb{R}^{kD_h\times D}$

```python
U_msa = torch.randn(K*D_h,D)
U_msa.size()
```

##### Calculate the output 

At first, we transpose dimentions of 2 and 3  of $\text{SA}$ (from $(B \times K \times (N+1) \times D_h)$ to $(B \times (N+1) \times K \times D_h)$). Then, we flatten the matrix starting dimention of 3. Afterwards, we can multiply it to $U_{msa}$.

```python
z_1_prime = torch.matmul(SA.transpose(1,2).flatten(2),U_msa) + z_0
z_1_prime.size()
```

##### Now, we put things together and define `MSA` class of our model.

We also add a dropout to calculate output of this module

```python
class MSA(nn.Module):
    
    def __init__(self, D, P, K, N,dropout_rate):
        super(MSA, self).__init__()
        self.N = N
        self.K = K
        self.D_h = int(D/K)
        self.U_qkv = nn.Parameter(torch.randn(D, K * 3 * self.D_h))
        self.U_msa = nn.Parameter(torch.randn(K*self.D_h,D))
        self.softmax = nn.Softmax(dim=-1)
        self.dropout = nn.Dropout(dropout_rate)
        self.norm = nn.LayerNorm(D)
    
    def forward(self,x):
        x = self.norm(x)
        B = x.shape[0]
        q,k,v = torch.split(torch.matmul(x, self.U_qkv).reshape(B,self.N,self.K,3*self.D_h).transpose(1,2), self.D_h, dim=3)
        A = self.softmax(torch.div(torch.matmul(q, k.transpose(2,3)),self.D_h ** -0.5))
        SA = torch.matmul(A, v)
        x = torch.matmul(SA.transpose(1,2).flatten(2),self.U_msa)
        x = self.dropout(x)
        return x
```

#### MLP

At first, we should do layer norm (LN) on the input. Then, depends on it is head MLP or MLP in a layer. At the end, we have residual connections after every MLP except for the Head MLP.

$z_{l} = \text{MLP}(\text{LN}(z_{l}^{\prime}))+z_{l}^{\prime}, \quad l=1,\dots,L$

```python
linear1 = nn.Linear(D,MLP_size)
activation = nn.GELU()
dropout = nn.Dropout(dropout_rate)
linear2 = nn.Linear(MLP_size,D)
```

We use GELU for activation function as described in the papar. Dropout rate is 0 in the paper in pre-training stage.

```python
x = linear1(z_1_prime)
x = activation(x)
x = dropout(x)
z_1 = linear2(x)
z_1.size()
```

```python
class MLP(nn.Module):
    
    def __init__(self, D, MLP_size,dropout_rate,N,num_classes=0,head=False):
        super(MLP, self).__init__()
        self.head = head
        self.linear1 = nn.Linear(D,MLP_size)
        self.activation = nn.GELU()
        self.dropout1 = nn.Dropout(dropout_rate)
        self.dropout2 = nn.Dropout(dropout_rate)
        self.linear2 = nn.Linear(MLP_size,D)
        self.norm = nn.LayerNorm(D)
        self.linear3 = nn.Linear(D,num_classes)
        
    def forward(self,x):
        x = self.norm(x)
        if self.head:
            x = self.linear3(x)
        else:
            x = self.linear1(x)
            x = self.activation(x)
            x = self.dropout1(x)
            x = self.linear2(x)
            x = self.dropout2(x)
        return x
```

### The final class of `Model`

We create model based on parameters. Inputs indicate that depth, number of classes, size of MLP, embedding size, size of batches, number of chunks of MSA, rate of dropout, and input images dimentions.

Thus, at first, `Embedding` class takes images as input. Then, in a loop, we put MSA, a residual connection, MLP, and another residual connection. At the end, we put Head MLP. 

```python
class Model(nn.Module):

    def __init__(self, D, P, K, MLP_size, image_size, depth, num_classes,dropout_rate,device):
        super(Model, self).__init__()
        H, W, C = image_size
        N = int(H*W/(P*P)) + 1
        self.depth = depth
        self.MLPs = []
        self.MSAs = []
        for i in range(depth):
            self.MLPs.append(MLP(D, MLP_size, dropout_rate,N).to(device))
            self.MSAs.append(MSA(D, P, K, N,dropout_rate).to(device))
        self.embedding = Embedding(D, P, N, C, B,dropout_rate).to(device)
        self.norm = nn.LayerNorm(D)
        self.MLP_head = MLP(D, MLP_size, dropout_rate,N,num_classes, True).to(device)

    def forward(self, x):
        x = self.embedding(x)
        for i in range(self.depth):
            x = self.MSAs[i](x) + x
            x = self.MLPs[i](x) + x
        x= self.MLP_head(x[:,0])
        return x
```

## Training

```python
model = Model(D,P,K,MLP_size,(H,W,C),depth,num_classes,dropout_rate,device)
model = model.to(device)
```

#### Define a function for calculating accuracy

```python
def calculate_accuracy(outputs,labels):
    _, predicted = torch.max(outputs.data, 1)
    total = labels.size(0)
    correct = (predicted == labels).sum().item()
    return correct/total
```

```python
train_loss_l = []
train_accuracy_l = []
num_epochs = 20

val_loss_l = []
val_accuracy_l = []

val_best=-10

for epoch in range(num_epochs):
    with tqdm(train_loader, unit="batch") as tepoch:
        tepoch.set_description('Epoch '+str(epoch+1))
        
        train_loss = 0
        train_accuracy = 0
        val_accuracy = 0
        val_loss = 0
        
        for images, labels in tepoch:

            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            
            train_loss += loss.item()
            train_accuracy += calculate_accuracy(outputs, labels)
            
            torch.cuda.empty_cache()
            del images, labels, outputs
        train_loss = train_loss/len(tepoch)
        train_accuracy = train_accuracy/len(tepoch)
        
        # Validation
        with torch.no_grad():
            correct = 0
            total = 0
            for images, labels in validation_loader:
                images = images.to(device)
                labels = labels.to(device)
                outputs = model(images)
                val_loss += criterion(outputs, labels).item()
                val_accuracy += calculate_accuracy(outputs, labels)
                del images, labels, outputs
                
            val_accuracy = val_accuracy / len(validation_loader)
            val_loss = val_loss / len(validation_loader)
        
        tepoch.close()    
        print('Train loss: ', train_loss, ' - Train accuracy: ', train_accuracy,' - Val loss: ', val_loss, ' - Val accuracy: ', val_accuracy)
        
                    
        train_accuracy_l.append(train_accuracy)
        train_loss_l.append(train_loss)
        
        val_accuracy_l.append(val_accuracy)
        val_loss_l.append(val_loss)
```

#### Save train and val loss and accuracies

```python
with open('log/train_loss_20.json', 'w') as outfile:
    json.dump(train_loss_l, outfile)
with open('log/val_loss_20.json', 'w') as outfile:
    json.dump(val_loss_l, outfile)
with open('log/train_accuracy_20.json', 'w') as outfile:
    json.dump(train_accuracy_l, outfile)
with open('log/val_accuracy_20.json', 'w') as outfile:
    json.dump(val_accuracy_l, outfile)
```

The model's accuracy is smoothly increasing. Although it seems model is overfitted, it needs more batches. If we had a model pretrained on larger dataset, model would be converged very sooner. `ViT` needs a lot of data for producing a good performance. As in this [page](https://wandb.ai/arutema47/cifar10-challange/reports/Untitled-Report--VmlldzoxNjU3MTU2?accessToken=3y3ib62e8b9ed2m2zb22dze8955fwuhljl5l4po1d5a3u9b7yzek1tz7a0d4i57r) displayed (from this [GitHub repository](https://github.com/kentaroy47/vision-transformers-cifar10/tree/712362493535517a21c2bbc9b8589374ddad8724)), a `ViT` model needs to trained for `200` epochs in order to learn `CIFAR-10`, and after $20$ epochs, the accuracy was $60\%$. Our dataset is similar to `CIFAR-10` in respect of scale. Or, this [page](https://keras.io/examples/vision/image_classification_with_vision_transformer/) shows similar problem when they train a `ViT` model on a small dataset. The other thing that we can conclude is that pre-training is very important aspect in using `ViT`. For example, in this [notebook](https://github.com/NielsRogge/Transformers-Tutorials/blob/master/VisionTransformer/Fine_tuning_the_Vision_Transformer_on_CIFAR_10_with_the_%F0%9F%A4%97_Trainer.ipynb), when they finetuned a pre-trained `ViT` model on `CIFAR-10` dataset, the accuracy of the model became $96\%$ only after $3$ epochs! I could not use pre-trained weight of `ViT` model because my implementation was a little bit different from available pre-trained models.

## Conclusions

* EuroSAT consists of $10$ classes with $27,000$ labeled and geo-referenced samples, which are approximately balanced.
* Pre-training is very important when we want to use `ViT` model. Fine-tuning a pre-trained model on a small dataset such `EuroSAT` can save lots of time and produce a very good performance. In general, `ViT` is not a good option for small datasets specifically when you want to train it from scratch. 
* Based on the confusion matrix and PCA, we can conclude that model did not train well. For example, although latent vectors of `Forest` and `SeaLake` were completely different from other labels, they are close to each other. Thus, the model could not distinguish between them as displayed in the confusion matrix.
